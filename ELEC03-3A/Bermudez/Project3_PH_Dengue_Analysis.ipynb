```python
Regional_Death = dengue.groupby(["Month"])[['Dengue_Deaths', 'Dengue_Cases']].sum().reset_index()
```

```python
Regional_Death
```

```python
Regional_Death["Mortality_Rate"] = np.round((Regional_Death["Dengue_Deaths"] / Regional_Death["Dengue_Cases"]) * 100,2)
```

```python
Regional_Death
```

```python
visual2 = sns.barplot(data=Regional_Death, x="Month",y="Mortality_Rate")
plt.gcf().set_size_inches(20,11)
plt.show()
```
